```python
import torch
import torch.nn as nn
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.tail()
```

```python
df.drop(columns=["id","Unnamed: 32"],inplace=True)
```

df.drop(columns=["id","Unnamed: 32"],inplace=True)

```python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

```python
x_train
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
y_train
```

```python
enabler=LabelEncoder()
y_train=enabler.fit_transform(y_train)
y_test=enabler.transform(y_test)
```

```python
y_train
```

```python
y_train.shape
```

```python
x_train_tensor=torch.from_numpy(x_train)
x_test_tensor=torch.from_numpy(x_test)
y_train_tensor=torch.from_numpy(y_train)
y_test_tensor=torch.from_numpy(y_test)
```

```python
x_train_tensor=x_train_tensor.to(torch.float32)
x_test_tensor=x_test_tensor.to(torch.float32)
y_train_tensor=y_train_tensor.to(torch.float32)
y_test_tensor=y_test_tensor.to(torch.float32)
```

```python
y_train_tensor.shape
```

```python
y_train_tensor.shape
```

```python
class Model(nn.Module):
  def __init__(self,num_features):
    super().__init__()
    self.linear=nn.Linear(num_features,1)
    self.sigmoid=nn.Sigmoid()
  def forward(self,x):
    out=self.linear(x)
    out=self.sigmoid(out)
    return out
```

```python
model.linear.weight.dtype
```

```python
model=Model(x_train_tensor.shape[1])
```

```python
learning_rate=0.01
epochs=100
lossfunc=nn.BCELoss()

```

```python
for epoch in range(epochs):
  y_pred=model(x_train_tensor)
  loss=lossfunc(y_pred,y_train_tensor.view(-1,1))
  optimizer=torch.optim.Adam(model.parameters(),lr=learning_rate)
  loss.backward()
  optimizer.step()
  optimizer.zero_grad()
  print(f'epoch:{epoch},loss:{loss}')
```

```python
y_pred_test=model(x_test_tensor)
y_pred_test
```

```python
y_pred_test_final=(y_pred_test>0.5).float()
y_pred_test_final
```

```python
accuracy=(y_pred_test_final==y_test_tensor).float().mean()
accuracy
print(f'Accuracy:{accuracy*100}%')
```
